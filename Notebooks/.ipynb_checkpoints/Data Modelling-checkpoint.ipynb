```python
import os
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report
import joblib

import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## Read data from the file

```python
file_path = os.path.join(os.getcwd(), '..', 'Data','2_PreProcessed.csv')
df = pd.read_csv(file_path)
```

```python
df['review'] = df['review'].astype(str)
```

```python
df.head()
```

## Vectorization

```python
tfidf_vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = tfidf_vectorizer.fit_transform(df['review'])
y = df['sentiment']
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8980)
```

## Model Training

```python
svm_model = LinearSVC()
svm_model.fit(X_train, y_train)
```

## Evaluation on Test

```python
y_pred_svm = svm_model.predict(X_test)
```

```python
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {accuracy_svm}")
print("SVM Classification Report:\n", classification_report(y_test, y_pred_svm))
```

```python
folder_path = os.path.join(os.getcwd(), '..', 'Model')
os.makedirs(folder_path, exist_ok=True)

model_path = os.path.join(folder_path,'sentiment_model.pkl')
joblib.dump(svm_model, model_path)
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Your new review
# new_review = "I love the shoes and the color of them is so good"
new_review = ""

# Preprocess the new review
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Preprocess the new review
preprocessed_review = preprocess_text(new_review)


new_review_tfidf = tfidf_vectorizer.transform([preprocessed_review])

# Predict sentiment using your SVM model
predicted_sentiment = svm_model.predict(new_review_tfidf)

# Interpret the predicted sentiment
if predicted_sentiment[0] == 0:
    decoded_sentiment = 'Negative'
else:
    decoded_sentiment = 'Positive'
print("Predicted Sentiment:", decoded_sentiment)

# Generate a word cloud for the new review
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(preprocessed_review)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title(f"Word Cloud for New Review ({decoded_sentiment} Sentiment)")
plt.axis('off')
plt.show()
```
